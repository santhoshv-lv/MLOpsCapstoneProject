Step 1: Data Ingestion and Cleaning:

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

try:
    # Step 1: Data Ingestion and Cleaning
    df = pd.read_csv('customer_shopping_data.csv')
    print("Dataset loaded successfully.")

    # Convert the 'invoice_date' column to datetime objects with dayfirst=True
    df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

    print("\nFirst 5 rows of the dataset:")
    print(df.head())

    print("\nInformation about the dataset:")
    df.info()

    # Step 2: Feature Engineering and Analysis
    df['total_price'] = df['quantity'] * df['price']
    print("\nAdded 'total_price' column.")

    df['invoice_year'] = df['invoice_date'].dt.year
    df['invoice_month'] = df['invoice_date'].dt.month
    df['invoice_quarter'] = df['invoice_date'].dt.quarter
    print("Added 'invoice_year', 'invoice_month', and 'invoice_quarter' columns.")

    most_recent_date = df['invoice_date'].max()
    rfm_df = df.groupby('customer_id').agg(
        recency=('invoice_date', lambda x: (most_recent_date - x.max()).days),
        frequency=('invoice_no', 'count'),
        monetary=('total_price', 'sum')
    ).reset_index()

    print("\nFirst 5 rows of the RFM DataFrame:")
    print(rfm_df.head())

    print("\nUpdated DataFrame info with new features:")
    df.info()

    # Step 3: Customer Segmentation using Machine Learning
    scaler = StandardScaler()
    rfm_scaled = scaler.fit_transform(rfm_df[['recency', 'frequency', 'monetary']])

    sse = []
    k_range = range(1, 11)
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
        kmeans.fit(rfm_scaled)
        sse.append(kmeans.inertia_)

    plt.figure(figsize=(10, 6))
    plt.plot(k_range, sse, marker='o')
    plt.title('The Elbow Method for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Sum of Squared Errors (SSE)')
    plt.grid(True)
    plt.savefig('elbow_method.png')
    print("\nGenerated 'elbow_method.png'")

    optimal_k = 3
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
    rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)

    print("\nFirst 5 rows of RFM DataFrame with cluster labels:")
    print(rfm_df.head())

except FileNotFoundError:
    print("Error: The file 'customer_shopping_data.csv' was not found. Please ensure the file is in the correct directory.")
```

```python
import pandas as pd

# Load the dataset
# Ensure the file 'customer_shopping_data.csv' is in the same directory as this script.
try:
    df = pd.read_csv('customer_shopping_data.csv')
    print("Dataset loaded successfully.")

    # Convert the 'invoice_date' column to datetime objects with dayfirst=True
    df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

    # Display the first 5 rows of the DataFrame
    print("\nFirst 5 rows of the dataset:")
    print(df.head())

    # Display a summary of the DataFrame including data types and non-null values
    print("\nInformation about the dataset:")
    df.info()

except FileNotFoundError:
    print("Error: The file 'customer_shopping_data.csv' was not found. Please ensure the file is in the correct directory.")
```

Step 2: Feature Engineering and Analysis

```python
import pandas as pd

# Assuming the DataFrame 'df' from the previous step is loaded

# 1. Feature Engineering: Calculate total price
df['total_price'] = df['quantity'] * df['price']
print("Added 'total_price' column.")

# 2. Feature Engineering: Extract time-series features
df['invoice_year'] = df['invoice_date'].dt.year
df['invoice_month'] = df['invoice_date'].dt.month
df['invoice_quarter'] = df['invoice_date'].dt.quarter
print("Added 'invoice_year', 'invoice_month', and 'invoice_quarter' columns.")

# 3. RFM Analysis: Calculate Recency, Frequency, and Monetary scores
# Find the most recent date in the dataset to calculate recency
most_recent_date = df['invoice_date'].max()

# Group by customer_id to compute RFM scores
rfm_df = df.groupby('customer_id').agg(
    recency=('invoice_date', lambda x: (most_recent_date - x.max()).days),
    frequency=('invoice_no', 'count'),
    monetary=('total_price', 'sum')
).reset_index()

# Display the first 5 rows of the new RFM DataFrame
print("\nFirst 5 rows of the RFM DataFrame:")
print(rfm_df.head())

# Display the updated info for the main DataFrame to show the new columns
print("\nUpdated DataFrame info with new features:")
df.info()
```

Step 3: Customer Segmentation using Machine Learning

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming rfm_df from the previous step is loaded

rfm_df = pd.DataFrame(rfm_df)

# Scale the RFM features
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df[['recency', 'frequency', 'monetary']])

# Use the Elbow Method to find the optimal number of clusters
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(rfm_scaled)
    sse.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o')
plt.title('The Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.grid(True)
plt.savefig('elbow_method.png')
print("Generated 'elbow_method.png'")

# Based on the elbow plot, choose a value for K (e.g., K=3 or K=4)
# For this example, let's assume K=3
optimal_k = 3

# Apply K-Means clustering with the chosen K
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Display the first 5 rows of the DataFrame with the new 'Cluster' column
print("\nFirst 5 rows of RFM DataFrame with cluster labels:")
print(rfm_df.head())
```
